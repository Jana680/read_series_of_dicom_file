```python
import pydicom 
import os
dc="/content/sample_data/dicom"
list1=[]
for i in os.listdir(dc):
    if i.endswith(".dcm"):
      list1.append(os.path.join(dc,i))
    

for i in list1:
  data=pydicom.dcmread(i)
  print(data)
```

```python
import matplotlib.pyplot as plt


print("Patient Name  : " , data.PatientName)
print("Patient ID  : " , data.PatientID)
print("Patient Sex : " , data.PatientSex)
print("Age  : " , data.PatientAge)

for i in list1:

  plt.imshow(data.pixel_array, cmap='gray')
  plt.show()
```
